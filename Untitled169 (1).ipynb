```python

conn = pyodbc.connect(
    r"DRIVER={ODBC Driver 17 for SQL Server};"
    r"SERVER=DESKTOP-GC78FEH\SQLEXPRESS;"
    r"DATABASE=AdventureWorks2019;" 
    r"Trusted_Connection=yes;"
    r"Encrypt=no;"
)

query_sales = """
SELECT 
    p.ProductID,
    p.Name AS ProductName,
    SUM(sod.OrderQty) AS TotalSold,
    SUM(sod.LineTotal) AS Revenue
FROM Production.Product p
JOIN Sales.SalesOrderDetail sod ON p.ProductID = sod.ProductID
GROUP BY p.ProductID, p.Name
ORDER BY TotalSold DESC
"""

df_sales = pd.read_sql(query_sales, conn)

conn.close()

print(df_sales.head())

```

```python
top10_sold = df_sales.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10_sold['ProductName'], top10_sold['TotalSold'], color='seagreen')
plt.xlabel("Total Sold")
plt.title("Top 10 Best-Selling Products")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python


conn = pyodbc.connect(
    r"DRIVER={ODBC Driver 17 for SQL Server};"
    r"SERVER=DESKTOP-GC78FEH\SQLEXPRESS;"
    r"DATABASE=AdventureWorks2019;"
    r"Trusted_Connection=yes;"
    r"Encrypt=no;"
)

query_color = """
SELECT 
    p.Color,
    SUM(sod.OrderQty) AS TotalSold
FROM Production.Product p
JOIN Sales.SalesOrderDetail sod ON p.ProductID = sod.ProductID
WHERE p.Color IS NOT NULL
GROUP BY p.Color
ORDER BY TotalSold DESC
"""

df_color = pd.read_sql(query_color, conn)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_color, x='TotalSold', y='Color', palette='viridis')
plt.xlabel('Total Sold')
plt.ylabel('Color')
plt.title('Total Products Sold by Color')
plt.tight_layout()
plt.show()
```

```python
query_size = """
SELECT 
    p.Size,
    SUM(sod.OrderQty) AS TotalSold
FROM Production.Product p
JOIN Sales.SalesOrderDetail sod ON p.ProductID = sod.ProductID
WHERE p.Size IS NOT NULL
GROUP BY p.Size
ORDER BY TotalSold DESC
"""

df_size = pd.read_sql(query_size, conn)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_size, x='TotalSold', y='Size', palette='coolwarm')
plt.xlabel('Total Sold')
plt.ylabel('Size')
plt.title('Total Products Sold by Size')
plt.tight_layout()
plt.show()
```

```python
df_weight_price = df[['Weight', 'ListPrice']].dropna()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_weight_price, x='Weight', y='ListPrice', color='darkblue')
plt.title('Relationship Between Weight and Price')
plt.xlabel('Weight')
plt.ylabel('List Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
